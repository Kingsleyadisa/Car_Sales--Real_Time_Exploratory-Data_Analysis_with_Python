# Car Sales Analysis using Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the csv file
file_path = 'Car_sales1.csv'
data = pd.read_csv(file_path)
```

```python
data
```

# **CAR SALES EXPLORATORY DATA ANALYSIS WITH PYTHON**

## **General**

```python
# 1. No. of rows and columns in the dataset
No_of_rows = len(data.index)
No_of_columns = len(data.columns)

print(f"No. of rows is: {No_of_rows}")
print(f"No. of columns is: {No_of_columns}")
```

```python
#2. Checking data types of each column
data.dtypes
```

```python
data
```

```python
#Changing the datatype of 'Latest_Launch' column from object to datatime object
data["Latest_Launch"] = pd.to_datetime(data["Latest_Launch"], format=('%m/%d/%Y'))
```

```python
# 3. Filling missing or null values

# a. counting how  many missing values in the data set
data.isna().sum()
```

```python
#b. Filling the missing values in the year_resale_value
average_ysv = data['__year_resale_value'].mean()
data['__year_resale_value'].replace(np.nan, average_ysv, inplace=True)

#data['year_resale_value'] = data['year_resale_value'].fillna(data['year_resale_value'].mean())
```

```python
data
```

```python
#c. Dropping other rows with null/missing values
data.dropna(inplace=True)
```

```python
data
```

```python
#4. Cleaning the the __year_resale_value column name
data.rename(columns={'__year_resale_value':'Year resale value'}, inplace=True)
```

```python
data
```

```python
#4b. Peforming descriptive statistics of the numerical columns
data.describe()
```

```python
#OR
#Statistical Analysis
median_sources = data[['Sales_in_thousands','Year resale value',	'Price_in_thousands','Engine_size','Horsepower','Wheelbase', 'Width',	'Length','Curb_weight','Fuel_capacity','Fuel_efficiency','Power_perf_factor']].median()
mean_sources = data[['Sales_in_thousands','Year resale value',	'Price_in_thousands','Engine_size','Horsepower','Wheelbase', 'Width',	'Length','Curb_weight','Fuel_capacity','Fuel_efficiency','Power_perf_factor']].mean()
std_sources = data[['Sales_in_thousands','Year resale value',	'Price_in_thousands','Engine_size','Horsepower','Wheelbase', 'Width',	'Length','Curb_weight','Fuel_capacity','Fuel_efficiency','Power_perf_factor']].std()
```

```python
S_analysis = pd.DataFrame([mean_sources, median_sources, std_sources], index=['Mean', 'Median', 'Standard Deviation'])
```

```python
S_analysis
```

## **Specific Analysis**

```python
#6. Manufacturer with the highest and lowest average sales volume
#a. Highest
data.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmax()
```

```python
#b. Smallest
data.groupby('Manufacturer')['Sales_in_thousands'].mean().idxmin()
```

```python
#7. distribution of car prices in the dataset
pivot1 = data.pivot_table(index='Manufacturer', values='Price_in_thousands', aggfunc='sum')
pivot1

pivot1.plot(kind='bar', figsize=(8, 5))
plt.xlabel('Manufacturer')
plt.ylabel('Sum of prices')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
#8. Correlation matrix of numerical variables in the dataset
data_correlation = data.corr()
data_correlation
```

```python
#Plotting the correlation
plt.figure(figsize=(10, 8))
sns.heatmap(data_correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#9. Plotting the relationship between price and Latest_Launch Year
sns.lineplot(x="Latest_Launch", y="Price_in_thousands", data=data)
plt.show()
```

```python
#10. Geouping the cars by manufacturers and analyzing their average price and fuel efficiency
Car_analysis = data.groupby('Manufacturer', as_index=False).agg({'Price_in_thousands':'mean', 'Fuel_efficiency':'count'})
```

```python
Car_analysis
```

```python
#11. Identifying and Visualizing Most Popular car brands

#a. Identifying
Most_popular = pd.DataFrame(data.groupby(['Manufacturer','Model'])['Sales_in_thousands'].max().sort_values(ascending=False))
```

```python
Most_popular
```

##### Therefore, the **Ford F-series** car is the Most Popular car

```python
#b. Visualizing the most popular car brands
Car_brands = pd.DataFrame(data.groupby('Manufacturer')['Sales_in_thousands'].max().sort_values(ascending=False))
Car_brands.plot(kind='bar', figsize=(12,7))
plt.title('Distribution of Most Popular Car Brands')
plt.show()
```

```python
#12. Car model with highest resale value compared to it's initial price
#model = data.pivot_table(index='Year resale value', columns='Model', values='Price_in_thousands').max().sort_values(ascending=False)
model = data.groupby(['Model','Price_in_thousands'])['Year resale value'].max().sort_values(ascending=False)
```
